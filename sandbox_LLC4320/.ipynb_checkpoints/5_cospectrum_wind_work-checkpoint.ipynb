```python
# reference: https://github.com/JhonKevin3015/VHFs/blob/main/PYTHONcodes/SPECTRAL.py 
# https:// doi.org/10.1029/2020JC016544 
# https://agupubs.onlinelibrary.wiley.com/doi/10.1029/2018JC014453
# https://gmd.copernicus.org/articles/15/8041/2022/

def cospec_ab_xarray(A_xr, B_xr, dx, dy, dt, window_type='hann', beta=3.86):
    """
    Compute 3D co-spectrum between two xarray.DataArray fields: A and B
    A_xr, B_xr: DataArrays with dims (lon, lat, time)
    dx, dy, dt: Grid spacing (assumed constant)
    window_type: 'hann', 'kais', 'flat', or None
    """
    # Check dimensions
    dims = A_xr.dims
    if dims != B_xr.dims or dims != ('i', 'j', 'time'):
        raise ValueError("Input DataArrays must have dimensions (lon, lat, time)")

    # Convert to numpy arrays
    A = A_xr.values
    B = B_xr.values
    l1, l2, l3 = A.shape
    N = A.size

    # Frequency axes
    df1 = 1./(l1*dx)
    df2 = 1./(l2*dy)
    df3 = 1./(l3*dt)
    f1Ny = 1./(2*dx)
    f2Ny = 1./(2*dy)
    f3Ny = 1./(2*dt)
    f1 = np.fft.fftshift(np.fft.fftfreq(l1, d=dx))
    f2 = np.fft.fftshift(np.fft.fftfreq(l2, d=dy))
    f3 = np.fft.rfftfreq(l3, d=dt)

    # Create window
    if window_type in ['hann', 'kais', 'flat']:
        if window_type == 'hann':
            wx = np.hanning(l1) * 0 + 1
            wy = np.hanning(l2)
            wt = np.hanning(l3)
        elif window_type == 'kais':
            wx = np.kaiser(l1, beta)
            wy = np.kaiser(l2, beta)
            wt = np.kaiser(l3, beta)
        elif window_type == 'flat':
            wx = signal.windows.flattop(l1)
            wy = signal.windows.flattop(l2)
            wt = signal.windows.flattop(l3)
        window_s_2d = np.outer(wx, wy)  # shape (l1, l2)
        window_s = np.repeat(window_s_2d[:, :, np.newaxis], l3, axis=2)
        window_t = np.repeat(wt, l1*l2).reshape(l3, l2, l1).T
    else:
        window_s = np.ones_like(A)
        window_t = np.ones_like(A)

    # Detrend along time axis (optional but recommended)
    #A = detrend(A, axis=2, type='linear')
    #B = detrend(B, axis=2, type='linear')

    # Apply window and FFT
    Ahat = np.fft.rfftn(window_s * window_t * A)
    Bhat = np.fft.rfftn(window_s * window_t * B)

    # Parseval normalization
    Aabs = (np.abs(Ahat)**2)
    Babs = (np.abs(Bhat)**2)
    ratio_A = np.sum(A**2) / np.sum(Aabs) * N
    ratio_B = np.sum(B**2) / np.sum(Babs) * N

    # Compute cospectrum
    cospec = (Bhat * Ahat.conj()).real / ((l1*l2*l3)**2) / (df1 * df2 * df3)
    cospec *= ratio_A * ratio_B

    # fftshift in lon-lat (first two dims)
    cospec_shifted = np.fft.fftshift(cospec, axes=(0,1))

    # Output xarray
    cospec_xr = xr.DataArray(
        cospec_shifted,
        dims=('freq_lon', 'freq_lat', 'freq_time'),
        coords={
            'freq_lon':  f1,
            'freq_lat':  f2,
            'freq_time': f3
        },
        attrs={
            'long_name': '3D Co-spectrum',
            'units': f"{A_xr.attrs.get('units', '')} * {B_xr.attrs.get('units', '')}"
        }
    )
    return cospec_xr, df1, df2, df3
```

```python
def isotropic_spectrum_3D_kf(spec_3d, kx_name='k_lon', ky_name='k_lat', f_name='f_time', nbins=50):
    """
    Convert 3D (kx, ky, f) spectrum into 2D isotropic wavenumber-frequency spectrum.

    Parameters:
        spec_3d : xarray.DataArray
            3D spectral field with dims (kx, ky, f)
        kx_name : str
            Name of wavenumber-x coordinate
        ky_name : str
            Name of wavenumber-y coordinate
        f_name : str
            Name of frequency coordinate
        nbins : int
            Number of isotropic wavenumber bins

    Returns:
        iso_spec : xarray.DataArray
            2D spectrum as function of isotropic wavenumber and frequency
            dims: ['k_iso', f_name]
    """
    # Get 1D wavenumbers and frequency
    kx = spec_3d[kx_name].values
    ky = spec_3d[ky_name].values
    f = spec_3d[f_name].values

    dkx = np.abs(kx[1] - kx[0])
    dky = np.abs(ky[1] - ky[0])

    KX, KY = np.meshgrid(kx, ky, indexing='ij')  # shape: (len(kx), len(ky))
    Kmag = np.sqrt(KX**2 + KY**2)

    kmax = Kmag.max()
    kbins = np.linspace(0, kmax, nbins + 1)
    k_iso = 0.5 * (kbins[:-1] + kbins[1:])

    # Get raw data
    spec_vals = spec_3d.values  # shape: (kx, ky, f)
    nf = spec_vals.shape[2]

    # Prepare output array
    iso_spec = np.full((nbins, nf), np.nan)

    for i in range(nbins):
        mask = (Kmag >= kbins[i]) & (Kmag < kbins[i+1])
        if not np.any(mask):
            continue
        for j in range(nf):
            selected = spec_vals[:, :, j][mask]
            iso_spec[i, j] = selected.mean()

    # Construct output xarray
    iso_spec_xr = xr.DataArray(
        iso_spec,
        dims=['k_iso', f_name],
        coords={'k_iso': k_iso, f_name: f},
        name=f'isotropic_{spec_3d.name}' if spec_3d.name else 'isotropic_spectrum',
        attrs={'description': 'Isotropic wavenumber-frequency spectrum'}
    )

    return iso_spec_xr
```

```python
u_surface_ij = gridx.interp(ds.SSU,'X')
v_surface_ij = gridx.interp(ds.SSV,'Y')
tau_x_ij = gridx.interp(ds.oceTAUX,'X')
tau_y_ij = gridx.interp(ds.oceTAUY,'Y')
```

```python
u_surface_ij = u_surface_ij.assign_coords(lat=u_surface_ij.j) 
u_surface_ij = u_surface_ij.assign_coords(lon=u_surface_ij.i) 
v_surface_ij = v_surface_ij.assign_coords(lat=v_surface_ij.j) 
v_surface_ij = v_surface_ij.assign_coords(lon=v_surface_ij.i) 

tau_x_ij = tau_x_ij.assign_coords(lat=tau_x_ij.j) 
tau_x_ij = tau_x_ij.assign_coords(lon=tau_x_ij.i) 
tau_y_ij = tau_y_ij.assign_coords(lat=tau_y_ij.j) 
tau_y_ij = tau_y_ij.assign_coords(lon=tau_y_ij.i) 
```

```python
A_xr = u_surface_ij
A_xr_detrended = A_xr - A_xr.mean(dim='time')          # 去掉时间平均
A_xr_detrended = A_xr_detrended - A_xr_detrended.mean(dim=('i', 'j'))  # 去掉每个时刻空间平均
u_surface_ij = A_xr_detrended                      
```

```python
A_xr = v_surface_ij
A_xr_detrended = A_xr - A_xr.mean(dim='time')          # 去掉时间平均
A_xr_detrended = A_xr_detrended - A_xr_detrended.mean(dim=('i', 'j'))  # 去掉每个时刻空间平均
v_surface_ij = A_xr_detrended  
```

```python
A_xr = tau_x_ij
A_xr_detrended = A_xr - A_xr.mean(dim='time')          # 去掉时间平均
A_xr_detrended = A_xr_detrended - A_xr_detrended.mean(dim=('i', 'j'))  # 去掉每个时刻空间平均
taux_surface_ij = A_xr_detrended  
```

```python
A_xr = tau_y_ij
A_xr_detrended = A_xr - A_xr.mean(dim='time')          # 去掉时间平均
A_xr_detrended = A_xr_detrended - A_xr_detrended.mean(dim=('i', 'j'))  # 去掉每个时刻空间平均
tauy_surface_ij = A_xr_detrended  
```

```python
OND = slice('2011-11-23','2011-12-31')
JFM = slice('2012-01-01','2012-03-31')
AMJ = slice('2012-04-01','2012-06-30')
JAS = slice('2012-07-01','2012-09-30')
```

```python
cospec_x, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=u_surface_ij.sel(time=OND).transpose('i', 'j', 'time'), 
                                                     B_xr=taux_surface_ij.sel(time=OND).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_y, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=v_surface_ij.sel(time=OND).transpose('i', 'j', 'time'), 
                                                     B_xr=tauy_surface_ij.sel(time=OND).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_OND = (cospec_x + cospec_y)/2
```

```python
cospec_x, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=u_surface_ij.sel(time=JFM).transpose('i', 'j', 'time'), 
                                                     B_xr=taux_surface_ij.sel(time=JFM).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_y, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=v_surface_ij.sel(time=JFM).transpose('i', 'j', 'time'), 
                                                     B_xr=tauy_surface_ij.sel(time=JFM).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_JFM = (cospec_x + cospec_y)/2
```

```python
cospec_x, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=u_surface_ij.sel(time=AMJ).transpose('i', 'j', 'time'), 
                                                     B_xr=taux_surface_ij.sel(time=AMJ).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_y, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=v_surface_ij.sel(time=AMJ).transpose('i', 'j', 'time'), 
                                                     B_xr=tauy_surface_ij.sel(time=AMJ).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_AMJ = (cospec_x + cospec_y)/2
```

```python
cospec_x, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=u_surface_ij.sel(time=JAS).transpose('i', 'j', 'time'), 
                                                     B_xr=taux_surface_ij.sel(time=JAS).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_y, df_lon, df_lat, df_time = cospec_ab_xarray(A_xr=v_surface_ij.sel(time=JAS).transpose('i', 'j', 'time'), 
                                                     B_xr=tauy_surface_ij.sel(time=JAS).transpose('i', 'j', 'time'), dx=2e3, dy=2e3, dt=3600, window_type='hann')
cospec_JAS = (cospec_x + cospec_y)/2
```

```python
# 输入数据是 cospec(k_lon, k_lat, f_time)，从 cospec_ab_xarray 得到
# 输出数据是 wavenumner-frequency cospec
cospec_kf_OND = isotropic_spectrum_3D_kf(cospec_OND, kx_name='freq_lon', ky_name='freq_lat', f_name='freq_time', nbins=100)
cospec_kf_JFM = isotropic_spectrum_3D_kf(cospec_JFM, kx_name='freq_lon', ky_name='freq_lat', f_name='freq_time', nbins=100)
cospec_kf_AMJ = isotropic_spectrum_3D_kf(cospec_AMJ, kx_name='freq_lon', ky_name='freq_lat', f_name='freq_time', nbins=100)
cospec_kf_JAS = isotropic_spectrum_3D_kf(cospec_JAS, kx_name='freq_lon', ky_name='freq_lat', f_name='freq_time', nbins=100)
```

```python
cospec_kf_OND['freq_time'].values[0] = 1e-8
cospec_kf_JFM['freq_time'].values[0] = 1e-8
cospec_kf_AMJ['freq_time'].values[0] = 1e-8
cospec_kf_JAS['freq_time'].values[0] = 1e-8
```

```python
cospec_kf_OND = cospec_kf_OND.assign_coords(freq_cph=cospec_kf_OND.freq_time*3600) 
cospec_kf_OND = cospec_kf_OND.assign_coords(k_cpkm=cospec_kf_OND.k_iso*2000) 
cospec_kf_JFM = cospec_kf_JFM.assign_coords(freq_cph=cospec_kf_JFM.freq_time*3600) 
cospec_kf_JFM = cospec_kf_JFM.assign_coords(k_cpkm=cospec_kf_JFM.k_iso*2000) 
cospec_kf_AMJ = cospec_kf_AMJ.assign_coords(freq_cph=cospec_kf_AMJ.freq_time*3600) 
cospec_kf_AMJ = cospec_kf_AMJ.assign_coords(k_cpkm=cospec_kf_AMJ.k_iso*2000) 
cospec_kf_JAS = cospec_kf_JAS.assign_coords(freq_cph=cospec_kf_JAS.freq_time*3600) 
cospec_kf_JAS = cospec_kf_JAS.assign_coords(k_cpkm=cospec_kf_JAS.k_iso*2000) 
```
