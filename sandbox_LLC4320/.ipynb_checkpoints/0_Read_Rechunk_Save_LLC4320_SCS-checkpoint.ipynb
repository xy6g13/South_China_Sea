# Read data

```python
LLC_data_folder = '/meddy/yuxlong5/data/LLC4320_SCS/llc_raw/'
ds_0 = xr.open_zarr(LLC_data_folder+'llc_0.zarr').drop(['Depth','CS','SN','PHrefC','PHrefF','dxG','dyG','dxC','dyC','rAw','rAs','rAz','niter','rA','k_l','k_p1','k_u','Z','Zl','Zp1','Zu','drC','drF'])
ds_1 = xr.open_zarr(LLC_data_folder+'llc_1.zarr').drop(['Depth','CS','SN','PHrefC','PHrefF','dxG','dyG','dxC','dyC','rAw','rAs','rAz','niter','rA','k_l','k_p1','k_u','Z','Zl','Zp1','Zu','drC','drF'])
ds_2 = xr.open_zarr(LLC_data_folder+'llc_2.zarr').drop(['Depth','CS','SN','PHrefC','PHrefF','dxG','dyG','dxC','dyC','rAw','rAs','rAz','niter','rA','k_l','k_p1','k_u','Z','Zl','Zp1','Zu','drC','drF'])
ds_3 = xr.open_zarr(LLC_data_folder+'llc_3.zarr').drop(['Depth','CS','SN','PHrefC','PHrefF','dxG','dyG','dxC','dyC','rAw','rAs','rAz','niter','rA','k_l','k_p1','k_u','Z','Zl','Zp1','Zu','drC','drF'])
ds = xr.concat([ds_0,ds_1,ds_2,ds_3], dim='time')
print(ds)
print('\n data size: %.1f GB' %(ds.nbytes / 1e9))
```

```python
lon = np.array(ds.XC.isel(j=1), dtype='>f')
lon
```

```python
lon_g = np.array(ds.XG.isel(j_g=1), dtype='>f')
lon_g
```

```python
lat = np.array(ds.YC.isel(i=1), dtype='>f')
lat
```

```python
lat_g = np.array(ds.YG.isel(i_g=1), dtype='>f')
lat_g
```

```python
ds = ds.assign_coords(j=lat) 
ds = ds.assign_coords(i=lon) 
ds = ds.assign_coords(j_g=lat_g) 
ds = ds.assign_coords(i_g=lon_g) 
ds = ds.drop(['k'])
ds
```

```python
out_dir_zarr = '/meddy/yuxlong5/data/LLC4320_SCS/'

chunks = {'time': 1, 'j': 1501, 'j_g': 1501, 'i': 1501, 'i_g': 1501}
ds = ds.chunk(chunks)
#del ds[v].encoding['chunks']
print(ds)
file_out = out_dir_zarr+'/llc_total_rechunk.zarr'
%time ds.to_zarr(file_out, mode='w') 
```
